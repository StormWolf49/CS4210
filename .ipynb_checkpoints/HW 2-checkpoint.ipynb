```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
import seaborn
```

```python
data_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
```

```python
dataset = pd.read_csv(data_url)
```

```python
null_mask = dataset.isnull().any(axis=1)
null_rows = dataset[null_mask]
print(null_rows)
```

```python
del dataset['Cabin']
null_mask = dataset.isnull().any(axis=1)
null_rows = dataset[null_mask]
print(null_rows)
```

```python
del dataset['Age']
null_mask = dataset.isnull().any(axis=1)
null_rows = dataset[null_mask]
print(null_rows)
```

```python
dataset.describe()
```

```python
dataset = dataset.dropna(how='any', axis=0)
null_mask = dataset.isnull().any(axis=1)
null_rows = dataset[null_mask]
print(null_rows)
```

```python
dataset.describe()
```

```python
plt.boxplot(dataset["Fare"])
```

```python
dataset = dataset[dataset["Fare"] < 400]
plt.boxplot(dataset["Fare"])
```
